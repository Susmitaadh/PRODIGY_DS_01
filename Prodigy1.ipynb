```python
df_clean = df[['Country/Territory', '2022 Population']].dropna()
```

```python
top10 = df_clean.sort_values(by='2022 Population', ascending=False).head(10)

plt.figure()
plt.bar(top10['Country/Territory'], top10['2022 Population'])
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2022)")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure()
plt.hist(df_clean['2022 Population'], bins=10)
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population Distribution (2022)")

plt.show()
```
